```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

file_path = r"C:\Users\5608t\OneDrive\桌面\digital_diet_mental_health.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
df.info()
```

```python
df.drop_duplicates()
```

```python
df.describe()
```

```python
number_cols = df.select_dtypes(include=["int64", "float64"]).columns.tolist()
plt.figure(figsize=(16,10))
sns.heatmap(df[number_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
scatter_cols = ["daily_screen_time_hours","sleep_quality","mood_rating","stress_level","physical_activity_hours_per_week","mental_health_score","weekly_depression_score"]
sns.pairplot(data=df[scatter_cols])
plt.tight_layout()
plt.show()
```

```python
spend_time_hours = ["social_media_hours","work_related_hours","entertainment_hours","gaming_hours","sleep_duration_hours"]
plt.pie(df[spend_time_hours].mean(), labels=spend_time_hours, autopct='%1.0f%%',startangle=90,shadow=True)
plt.axis('equal')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
plt.figure(figsize=(12,6))
sns.boxplot(x="age",y="daily_screen_time_hours",data=df)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
usage_hours = ["phone_usage_hours","laptop_usage_hours","tv_usage_hours","tablet_usage_hours"]
sns.barplot(x= usage_hours,y = df[usage_hours].mean())
plt.title("average hour for device")
plt.xlabel("usage hours")
plt.xticks(rotation=45)
plt.ylabel("average usage hours")
plt.tight_layout()
plt.show()
```
